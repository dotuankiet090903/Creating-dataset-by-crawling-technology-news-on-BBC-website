```python
from transformers import pipeline

def summarize_text(inputs):
    summarizer = pipeline('summarization')

    summaries = []
    count = 1
    for input in inputs:
        # Use the pipeline to summarize the text
        if len(input) > 1024:
            input = input[:1024]
        summary = summarizer(input, max_length=100, min_length=1, do_sample=False)
        print("{} line".format(count))
        print("Source inputs (truncated if length < 1024): {}".format(input))
        print("Summarized version: {}".format(summary[0]['summary_text']))
        print("-"*300)
        summaries.append(summary[0]['summary_text'])
        count+=1
    return summaries


# Test the function
summaries = summarize_text(inputs[0:499])
print("There are {} documents to be summarized".format(len(summaries)))
```

```python
import numpy as np
content = df[' Content']
summaries = summaries + [np.nan]*(len(content) - len(summaries))
data = {'Content' : content, 'Summarize' : summaries}
new_df = pd.DataFrame(data)
```
