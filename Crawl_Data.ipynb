```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_links (url):
  response = requests.get(url)
  return response
```

```python
# URL for the BBC Technology section
response = get_links ("https://www.bbc.com/innovation")
# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

article_links = []
titles = []
content = []
```

```python
# Find all articles with the specified class and taking its article's link
for anchor in soup.find_all(attrs={"data-testid": "anchor-inner-wrapper"}):
  articles_link = soup.find_all('a', attrs={"data-testid": "internal-link"}, href=True)
  for links in articles_link:
    # Taking article link only
    if links['href'].startswith("/news/articles"):
      article_links.append(links['href'])
```

```python
# Taking titles of each article's link
for link in article_links:
  response = get_links("https://www.bbc.com" + link)
  soup = BeautifulSoup(response.content, "html.parser")
  for title in soup.find_all("h1"):
      title = title.text.strip()
      titles.append(title)

print(titles)
```

```python
# Taking content of each article's link
for link in article_links:
  response = get_links("https://www.bbc.com" + link)
  soup = BeautifulSoup(response.content, "html.parser")
  for contents in soup.find_all(attrs={"data-component": "text-block"}):
      contents = contents.text.strip()
      content.append(contents)
print(content)
```

```python
print(len(article_links))
print(len(titles))
print(len(content))
```

```python
import pandas as pd
temp = zip(titles, content)
t, c = [], []
for a, b in temp:
  t.append(a)
  c.append(b)
data = {'Title': t, 'Content': c}
df = pd.DataFrame(data)
df
```

```python
df.to_csv('data.csv')
```
